```python
def normalize(input_image, input_mask):
    input_image = tf.cast(input_image, tf.float32) / 255.0
    input_mask = tf.cast(input_mask, tf.float32) / 127.5
    return input_image, input_mask
```

```python
def load_image_and_mask(image_filename, mask_filename):
    
    image_string = tf.io.read_file(image_filename)
    image_decoded = tf.io.decode_jpeg(image_string,channels = 3)
    input_image = tf.image.resize(image_decoded, (160, 240))
    
    mask_string = tf.io.read_file(mask_filename)
    mask_decoded = tf.io.decode_jpeg(mask_string, channels = 3)
    mask_grayscale = tf.image.rgb_to_grayscale(mask_decoded)
    input_mask = tf.image.resize(mask_grayscale, (160, 240))

    input_image, input_mask = normalize(input_image, input_mask)

    return input_image, input_mask
```

```python
BATCH_SIZE = 32
BUFFER_SIZE = 1000
```

```python
full_dataset = tf.data.Dataset.from_tensor_slices((img_files, mask_files))
full_dataset = full_dataset.map(load_image_and_mask, num_parallel_calls=tf.data.AUTOTUNE)
full_dataset = full_dataset.batch(BATCH_SIZE)
```

```python
def display(display_list):
    
    plt.figure(figsize=(15, 15))

    title = ['Input Image', 'True Mask', 'Predicted Mask']

    for i in range(len(display_list)):
        plt.subplot(1, len(display_list), i+1)
        plt.title(title[i])
        plt.imshow(tf.keras.preprocessing.image.array_to_img(display_list[i]))
        plt.axis('off')
        
    plt.show()
```

### 2. Load Model

Here we load the model that we had saved after training

```python
model = tf.keras.models.load_model(drive_path + 'data/saved_model/240-160-CNN.h5')
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
def create_mask(pred_mask):
    pred_mask = tf.argmax(pred_mask, axis=-1)
    pred_mask = pred_mask[..., tf.newaxis]
    return pred_mask
```

```python
def show_predictions(dataset, num):

    for image, mask in dataset.take(num):
        
        pred_mask = model.predict(image)
        display([image[0], mask[0], create_mask(pred_mask[0])])
```

```python
show_predictions(full_dataset, 3)
```

Next we attempt to encode the predicted segmentation masks into Run-Length-Encoding format.

```python
def rle_encode(tf_img):

    img_array = np.array(tf_img)

    pixels = []

    for i in range(img_array.shape[0]):

        for j in range(img_array.shape[1]):

            if int(img_array[i][j][0]) != 0:
                
                pix = ((i * 1920) + j + 1)
                pixels.append(pix)

    pixel_breaks = [0]

    for i in range(len(pixels) - 1):

        if pixels[i] != ((pixels[i+1]) - 1):

            pixel_breaks.append(i+1)

    pixel_breaks.append(len(pixels))

    rle_mask = ''

    for i in range(len(pixel_breaks) - 1):

        rle_mask += (str(pixels[pixel_breaks[i]]) + ' ')
        rle_mask += (str(pixel_breaks[i+1] - pixel_breaks[i]) + ' ')

    return rle_mask[:-1]
```

```python
rle_encode(masks_resized_1000[0])
```

When attempting to encode 1000 images into RLE, the code doesn't complete even after 20+ minutes.

This shows that the function could be improved to be more efficient.
